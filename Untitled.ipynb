```python
import faiss
from sentence_transformers import SentenceTransformer
from sklearn.metrics.pairwise import cosine_similarity

# Load datasets
client_df = pd.read_csv("client.csv")
freelance_df = pd.read_csv("freelance.csv")

# Load pre-trained Sentence-BERT model
model = SentenceTransformer("sentence-transformers/all-MiniLM-L6-v2")

# Convert client job descriptions into embeddings
client_df["combined_text"] = client_df["Job Title"] + " " + client_df["Description"] + " " + client_df["Merged_Category"]
client_embeddings = model.encode(client_df["combined_text"].tolist(), convert_to_numpy=True)

# Convert freelancer profiles into embeddings
freelance_df["combined_text"] = freelance_df["job_skills"] + " " + freelance_df["experience"] + " " + freelance_df["projects"]
freelancer_embeddings = model.encode(freelance_df["combined_text"].tolist(), convert_to_numpy=True)

print(f"Generated {len(client_embeddings)} client embeddings and {len(freelancer_embeddings)} freelancer embeddings.")
```
